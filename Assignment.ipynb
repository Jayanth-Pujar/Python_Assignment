# Data Insights

```python
# Visualization 1: Monthly counts of complaints received and sent to company
df['Date received'] = pd.to_datetime(df['Date received'])
df['Date sent to company'] = pd.to_datetime(df['Date sent to company'])
df['Month received'] = df['Date received'].dt.to_period('M')
df['Month sent to company'] = df['Date sent to company'].dt.to_period('M')
monthly_counts_received = df.groupby('Month received').size()
monthly_counts_sent = df.groupby('Month sent to company').size()
plt.figure(figsize=(10, 6))
plt.plot(monthly_counts_received.index.to_timestamp(), monthly_counts_received.values, label='Complaints Received', linewidth=2.5)
plt.plot(monthly_counts_sent.index.to_timestamp(), monthly_counts_sent.values, label='Complaints Sent to Company', linewidth=2.5)
plt.xlabel('Date')
plt.ylabel('Number of Complaints')
plt.title('Monthly Complaints Received and Sent')
plt.legend(loc='upper left')
plt.show()
```

```python
# Visualization 2: Top N product categories by complaints
top_n = 10
top_products = df['Product'].value_counts().head(top_n)

plt.figure(figsize=(10, 6))
top_products.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('Number of Complaints')
plt.title(f'Top {top_n} Product Categories by Complaints')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 3: Top N companies by complaints
top_companies = df['Company'].value_counts().head(top_n)

plt.figure(figsize=(10, 6))
top_companies.plot(kind='bar')
plt.xlabel('Company')
plt.ylabel('Number of Complaints')
plt.title(f'Top {top_n} Companies by Complaints')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 4: Top N states by complaints
top_states = df['State'].value_counts().head(top_n)

plt.figure(figsize=(10, 6))
top_states.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Number of Complaints')
plt.title(f'Top {top_n} States by Complaints')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 5: Bar chart for distribution of responses to timely response
plt.figure(figsize=(6, 4))
df['Timely response?'].value_counts().plot(kind='bar')
plt.xlabel('Timely Response?')
plt.ylabel('Number of Complaints')
plt.title('Distribution of Timely Responses')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 6: Bar chart for distribution of consumer disputes
plt.figure(figsize=(6, 4))
df['Consumer disputed?'].value_counts().plot(kind='bar')
plt.xlabel('Consumer Disputed?')
plt.ylabel('Number of Complaints')
plt.title('Distribution of Consumer Disputes')
plt.xticks(rotation=45)
plt.show()
```

## 2. Given an unsorted array of integers, find the length of the longest continuous increasing subsequence (subarray). 

```python
# Input list of integers from the user
nums = []
n = int(input("Enter the number of elements: "))
for i in range(n):
    nums.append(int(input(f"Enter element {i + 1}: ")))

if not nums:
    print(0)
else:
    result = [1] * len(nums)

    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            result[i] = result[i - 1] + 1
    print(max(result))  
```

## 3. Given a list of non negative integers, arrange them such that they form the largest number.

```python
def largestNumber(array):
    array.sort(key=lambda x: str(x), reverse=True)
    result = ''.join(map(str, array))
    if result == '0' * len(result):
        return '0'
    else:
        return result

if __name__ == "__main__":
    n = int(input("Enter the number of elements: "))
    numbers = []
    for i in range(n):
        num = int(input(f"Enter number {i + 1}: "))
        numbers.append(num)
    
    print("The largest numberis:", largestNumber(numbers))
```

## 4. Store all the "servlet-name", and "servlet-class" to a csv file from the attached sample_json.json file using Python.

```python
import json
import csv

# loading json data
with open('sample_json.json') as json_file:
    data = json.load(json_file)

# extracting servlet-name and servlet-class
servlet_data = []
for servlet in data['web-app']['servlet']:
    servlet_name = servlet['servlet-name']
    servlet_class = servlet['servlet-class']
    servlet_data.append([servlet_name, servlet_class])

# creating a csv file
with open('servlet_data.csv', mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['servlet-name', 'servlet-class'])
    writer.writerows(servlet_data) 
```
